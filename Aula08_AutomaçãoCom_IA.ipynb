```python
#1 crie um script Python simples para verificar a temeratura de um sistem.Se for maior que 30 ,exiba um alerta de superaquecimento

# Bloco 'try' para lidar com possíveis erros de entrada do usuário.
try:
    # Solicita ao usuário que digite a temperatura atual e armazena a entrada como string.
    temperature_input = input("Por favor, digite a temperatura atual: ")
    # Converte a entrada do usuário (string) para um número inteiro.
    temperature = int(temperature_input)

    # Verifica se a temperatura é maior que 30 graus Celsius.
    if temperature > 30:
        # Se a temperatura for maior que 30, exibe uma mensagem de alerta de superaquecimento.
        print(f"ALERTA: Temperatura atual é {temperature}°C. Superaquecimento detectado!")
    # Se a temperatura não for maior que 30 (ou seja, for 30 ou menos).
    else:
        # Exibe uma mensagem informando que o sistema está operando normalmente.
        print(f"Temperatura atual é {temperature}°C. Sistema operando normalmente.")
# Bloco 'except ValueError' para capturar erros caso o usuário não digite um número inteiro.
except ValueError:
    # Se ocorrer um erro de conversão (ex: usuário digitou texto), exibe uma mensagem de erro.
    print("Entrada inválida. Por favor, digite um número inteiro para a temperatura.")
```

```python
#2 criar um programa a que organize arquivos bagunçados em pastas automaticamente, baseado-se apeans na extenção do arquivo (ex: pdf,jpg) use biblioteca os ou shutil

import os
import shutil

def organize_files(source_dir='./'):
    """
    Organiza arquivos em pastas baseadas em suas extensões.

    Args:
        source_dir (str): O diretório onde os arquivos bagunçados estão localizados.
                          O padrão é o diretório atual.
    """
    print(f"Iniciando organização de arquivos em: {os.path.abspath(source_dir)}")

    if not os.path.isdir(source_dir):
        print(f"Erro: O diretório de origem '{source_dir}' não existe.")
        return

    # Cria alguns arquivos de exemplo para teste (remova ou comente esta parte em uso real)
    if not os.path.exists(os.path.join(source_dir, 'documento.pdf')):
        with open(os.path.join(source_dir, 'documento.pdf'), 'w') as f: f.write('dummy pdf')
    if not os.path.exists(os.path.join(source_dir, 'imagem.jpg')):
        with open(os.path.join(source_dir, 'imagem.jpg'), 'w') as f: f.write('dummy jpg')
    if not os.path.exists(os.path.join(source_dir, 'relatorio.docx')):
        with open(os.path.join(source_dir, 'relatorio.docx'), 'w') as f: f.write('dummy docx')
    if not os.path.exists(os.path.join(source_dir, 'planilha.xlsx')):
        with open(os.path.join(source_dir, 'planilha.xlsx'), 'w') as f: f.write('dummy xlsx')
    if not os.path.exists(os.path.join(source_dir, 'programa.py')):
        with open(os.path.join(source_dir, 'programa.py'), 'w') as f: f.write('dummy py')
    print("Arquivos de exemplo criados para demonstração.")


    for filename in os.listdir(source_dir):
        # Ignora diretórios
        if os.path.isdir(os.path.join(source_dir, filename)):
            continue

        file_path = os.path.join(source_dir, filename)

        # Ignora arquivos sem extensão
        if '.' not in filename:
            print(f"Ignorando '{filename}': sem extensão conhecida.")
            continue

        # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
        file_extension = filename.split('.')[-1].lower()

        # Define o nome do diretório de destino (ex: 'pdf_files', 'jpg_files')
        destination_folder = os.path.join(source_dir, f'{file_extension}_files')

        # Cria o diretório de destino se ele não existir
        os.makedirs(destination_folder, exist_ok=True)

        # Move o arquivo para a pasta de destino
        try:
            shutil.move(file_path, os.path.join(destination_folder, filename))
            print(f"Movido: '{filename}' para '{destination_folder}'")
        except Exception as e:
            print(f"Erro ao mover '{filename}': {e}")

    print("Organização de arquivos concluída!")

# Para executar o programa, chame a função:
# organize_files() # Isso organizará arquivos no diretório atual

# Exemplo de como você pode chamar a função para organizar arquivos no diretório atual
organize_files()
```

```python
#3 desenvolver um script que consulte uma API publica (via CEp e retorne as informaçoes de endereço legivel. Apresnete o tratamento de script try/except)

import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço a partir de um CEP.

    Args:
        cep (str): O CEP (Código de Endereçamento Postal) a ser consultado.
                   Deve ser uma string de 8 dígitos (ex: '01001000').

    Returns:
        dict or None: Um dicionário com as informações do endereço se a consulta for bem-sucedida,
                      ou None se ocorrer um erro ou o CEP for inválido.
    """
    # Limpa o CEP, removendo hífens e outros caracteres não numéricos
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem 8 dígitos
    if len(cep) != 8:
        print(f"Erro: O CEP '{cep}' é inválido. Por favor, insira 8 dígitos numéricos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição HTTP GET para a API
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx)

        data = response.json()

        # Verifica se a API retornou um erro (CEP não encontrado, por exemplo)
        if 'erro' in data and data['erro']:
            print(f"Erro: CEP '{cep}' não encontrado.")
            return None

        return data

    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP ao consultar o CEP '{cep}': {http_err}")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de conexão ao consultar o CEP '{cep}': {conn_err}\nVerifique sua conexão com a internet.")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Tempo limite excedido ao consultar o CEP '{cep}': {timeout_err}\nO servidor demorou muito para responder.")
    except requests.exceptions.RequestException as req_err:
        print(f"Ocorreu um erro inesperado ao consultar o CEP '{cep}': {req_err}")
    except ValueError:
        print(f"Erro: Não foi possível decodificar a resposta JSON para o CEP '{cep}'.")
    except Exception as e:
        print(f"Ocorreu um erro desconhecido ao consultar o CEP '{cep}': {e}")

    return None

def exibir_endereco(dados_endereco):
    """
    Exibe as informações do endereço de forma legível.

    Args:
        dados_endereco (dict): Dicionário contendo as informações do endereço.
    """
    if dados_endereco:
        print("\n--- Informações do Endereço ---")
        print(f"CEP: {dados_endereco.get('cep', 'N/A')}")
        print(f"Logradouro: {dados_endereco.get('logradouro', 'N/A')}")
        print(f"Complemento: {dados_endereco.get('complemento', 'N/A')}")
        print(f"Bairro: {dados_endereco.get('bairro', 'N/A')}")
        print(f"Localidade: {dados_endereco.get('localidade', 'N/A')}")
        print(f"UF: {dados_endereco.get('uf', 'N/A')}")
        print(f"DDD: {dados_endereco.get('ddd', 'N/A')}")
        print("-------------------------------")
    else:
        print("Não foi possível obter informações de endereço.")

# --- Exemplo de uso ---
if __name__ == "__main__":
    # Exemplo de CEP válido
    cep_valido = input("Digite um CEP para consulta (ex: 01001-000): ")
    endereco = consultar_cep(cep_valido)
    exibir_endereco(endereco)

    print("\nTestando com um CEP inválido (formato incorreto)")
    cep_formato_errado = "12345"
    endereco_errado = consultar_cep(cep_formato_errado)
    exibir_endereco(endereco_errado)

    print("\nTestando com um CEP que não existe")
    cep_nao_existe = "99999999"
    endereco_nao_existe = consultar_cep(cep_nao_existe)
    exibir_endereco(endereco_nao_existe)
```
