```python
# Solicita os dois números ao usuário
num1_str = input("Digite o primeiro número: ")
num2_str = input("Digite o segundo número: ")

# Converte as entradas para float para permitir números decimais
try:
    num1 = float(num1_str)
    num2 = float(num2_str)
except ValueError:
    print("Entrada inválida. Por favor, digite apenas números.")
    exit()
```

```python
# Pede ao usuário para escolher a operação
operacao = input("Escolha a operação ('somar' ou 'subtrair'): ").lower()

# Realiza a operação e exibe o resultado
if operacao == 'somar':
    resultado = num1 + num2
    print(f"A soma é: {resultado}")
elif operacao == 'subtrair':
    resultado = num1 - num2
    print(f"A subtração é: {resultado}")
else:
    print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
```
